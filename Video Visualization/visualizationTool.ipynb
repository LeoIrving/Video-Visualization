```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pose_df = pd.read_csv("pose.csv")
```

```python
pose_df
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111,projection='3d')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
i = 1
for row in pose_df.iterrows():
    Color = [i*5/255,i*5/255,i*10/255]
    ax.scatter(row[1][0],row[1][1],row[1][2],s=10,color=Color, marker='x')
    i = i+ 0.1
    if i >25.5:
        i = 1
ax.view_init(-120, 60)
plt.show()
```
